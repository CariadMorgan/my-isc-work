Q1. Import the `'../data/xbhubo.pgc0apr.nc'` dataset and create the temperature data array as in the last lesson.

```python
import xarray as xr
ds = xr.load_dataset('~/my-isc-work/python-data/data/xbhubo.pgc0apr.nc')
temp = ds['temp']
```

Q2. Try a simple `.plot()` on your temperature data array, to see what xarray does. Why has it done this?

```python
temp.plot()
```

Q3. Next, try the same with a 2-dimensional view of your dataset. Try selecting sea surface temperature values and plotting those.

```python
temp.sel(depth=0, method='nearest').plot()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp.sel(latitude=0, method='nearest').plot()
ax.invert_yaxis()
ax.set_title('Depth profile of temperature at Equator')
```

Q4. Finally, create a depth profile from `temperature` by using `sel` to select data for the same latitude and longitude values (31,0).

Hint: The latitude value is `-50.625` and the longitude value is `0`. All 3 of these methods will return the same dataset.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp.sel(latitude=-50.624, longitude=0, method='nearest').plot()
ax.invert_yaxis()
ax.set_title('Depth profile at specific point defined')
```

Q5. Create a plot from this time series.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp.sel(latitude=-50.624, longitude=0, method='nearest').plot()
ax.invert_yaxis()
ax.set_title('Depth profile at specific point defined')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp.sel(latitude=-50.624, longitude=0, method='nearest').plot(color='red', marker = 'x' )
ax.invert_yaxis()
ax.set_title('Depth profile at specific point defined')
```

Q7. Create a time series plot comparing the temperature profile at three different grid cells:
 - lat = -50.625, lon = 0
 - lat = 0, lon = 200
 - lat = 0, lon = 320

Make sure each time series has a different colour and include a legend. As an extension, give them different linestyles too.
Hint: use `.sel` to index the lat and lon.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
temp.sel(latitude=-50.624, longitude=0, method='nearest').plot(color='red')
temp.sel(latitude=0, longitude=200, method='nearest').plot(color='blue')
temp.sel(latitude=0, longitude=320, method='nearest').plot(color='green')
ax.invert_yaxis()
ax.set_title('Depth profile at specific point defined')
```

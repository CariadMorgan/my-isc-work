## GroupBy processing

Our ocean model dataset has no time dimension, so for this exercise we are going to use the NOAA ERSSST dataset from the tutorial. Load it using the command below.

```python
import xarray as xr
ds = xr.tutorial.load_dataset("ersstv5")
```

Have a quick explore of the dataset and see what it contains.

```python
ds
```

Q1. First, lets group our dataset by year.

```python
# this .dt syntax to index the data by year, then this can be used to group the data
# ds.time.dt.year

# ds_gb_year = ds.groupby(ds.time.dt.year)

#or can do it this way

ds_gb_year = ds.groupby('time.year')
```

Q2. Lets take the mean of each group, to give the annual mean.

```python
yearly_mean = ds_gb_year.mean()
yearly_mean
```

Q3. To see what we've done, lets plot the mean for the year 1960.

```python
#yearly_mean.sel(year = 1970, method = 'nearest').sst.plot()

#different convention - this is better as if you have a dimensions with the same 
#name as a function - it won't impact your code

yearly_mean['sst'].sel(year = 1970, method = 'nearest').plot()
```

Q4. Next, lets plot an annual mean time series for the point in the Atlantic ocean latitude=-50.625, longitude=0.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
yearly_mean['sst'].sel(lat=-50.625, lon=0, method='nearest').plot()
ax.set_title('Annual mean time series for specific point')
```
